```python
import os
import random
import shutil

# Paths
BASE_DIR = "/content/data"
SOURCE_DIR = os.path.join(BASE_DIR, "videos_subset")

TRAIN_DIR = os.path.join(BASE_DIR, "train")
VAL_DIR = os.path.join(BASE_DIR, "val")
TEST_DIR = os.path.join(BASE_DIR, "test")

# Create folders
os.makedirs(TRAIN_DIR, exist_ok=True)
os.makedirs(VAL_DIR, exist_ok=True)
os.makedirs(TEST_DIR, exist_ok=True)

# Get video list
videos = sorted(os.listdir(SOURCE_DIR))
random.shuffle(videos)

# Split sizes
train_split = int(0.7 * len(videos))
val_split = int(0.85 * len(videos))

train_videos = videos[:train_split]
val_videos = videos[train_split:val_split]
test_videos = videos[val_split:]

# Copy files
def copy_files(video_list, dest):
    for v in video_list:
        shutil.copy(os.path.join(SOURCE_DIR, v), os.path.join(dest, v))

copy_files(train_videos, TRAIN_DIR)
copy_files(val_videos, VAL_DIR)
copy_files(test_videos, TEST_DIR)

print("Dataset split completed")
print("Train:", len(train_videos))
print("Validation:", len(val_videos))
print("Test:", len(test_videos))
```

```python
import json
import numpy as np
from rouge_score import rouge_scorer
from nltk.translate.meteor_score import meteor_score
from bert_score import score as bert_score
from sacrebleu import corpus_bleu
from pycocoevalcap.cider.cider import Cider
import nltk

nltk.download('wordnet', quiet=True)

# Load your file
with open("/content/VLM-Road_crash_work/results/zero-shot/detailed_results.json") as f:
    data = json.load(f)

preds = [x["prediction"] for x in data]
refs = [x["reference"] for x in data]

# BLEU
bleu = corpus_bleu(preds, [refs])
bleu_scores = {
    "bleu_1": bleu.precisions[0],
    "bleu_2": bleu.precisions[1],
    "bleu_3": bleu.precisions[2],
    "bleu_4": bleu.precisions[3],
}

# METEOR
meteor_vals = [meteor_score([r.split()], p.split()) for p, r in zip(preds, refs)]
mean_meteor = float(np.mean(meteor_vals))

# ROUGE
scorer = rouge_scorer.RougeScorer(["rouge1", "rouge2", "rougeL"], use_stemmer=True)
r1, r2, rl = [], [], []

for p, r in zip(preds, refs):
    s = scorer.score(r, p)
    r1.append(s["rouge1"].fmeasure)
    r2.append(s["rouge2"].fmeasure)
    rl.append(s["rougeL"].fmeasure)

# BERTScore
_, _, F1 = bert_score(preds, refs, lang="en", device="cpu")

# CIDEr
cider = Cider()
cider_vals = [cider.compute_score({0: [r]}, {0: [p]})[0] for p, r in zip(preds, refs)]

# Final JSON
results = {
    "num_samples": len(preds),
    "bleu_scores": bleu_scores,
    "meteor": mean_meteor,
    "rouge_1": float(np.mean(r1)),
    "rouge_2": float(np.mean(r2)),
    "rouge_l": float(np.mean(rl)),
    "bertscore": float(F1.mean()),
    "cider": float(np.mean(cider_vals)),
    "nli_scores": {
        "entailment_accuracy": 0.0,
        "avg_entailment_prob": 0.0,
        "contradiction_rate": 0.0,
        "neutral_rate": 0.0
    }
}

# Save
with open("results/zero-shot/metrics.json", "w") as f:
    json.dump(results, f, indent=2)

print("Zero-shot metrics saved ✅")
```
